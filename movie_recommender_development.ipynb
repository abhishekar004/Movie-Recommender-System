```python
movies= movies[['Poster', 'Title', 'Certificate', 'Genre', 'Rating', 'Director', 'Cast', 'Description']]
```

```python
movies['tags']= movies['Description']+ movies['Genre']
```

```python
movies.head()
```

```python
new_data= movies.drop(columns=['Cast','Director','Rating'])
```

```python
new_data.isnull().sum()
```

```python
new_data['Certificate']= new_data['Certificate'].fillna("PG-13")
```

```python
new_data.isnull().sum()
```

```python
null_data = new_data[new_data.isnull().any(axis=1)]
```

```python
new_data=new_data.dropna(subset=['Genre'])
```

```python
import faiss
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
tfidf= TfidfVectorizer(max_features=10000, stop_words='english')
vector = tfidf.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
vector = np.array(vector, dtype= 'float32')
index = faiss.IndexFlatL2(vector.shape[1])
index.add(vector)
```

```python
new_data[new_data['Title']=="Unfrosted"].index[0]
```

```python
def recommend_faiss(movie):
    try:
        movie_index = new_data[new_data['Title']==movie].index[0]
        D, I = index.search(vector[movie_index].reshape(1, -1,),6)
        recommended_movies = [new_data.iloc[idx].Title for idx in I[0] if idx != movie_index]
        return recommended_movies
    except Exception as e:
        return [f"Error: {e}"]
```

```python
recommend_faiss("Challengers")
```
